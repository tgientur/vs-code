## 1.2 Python访问MySQL数据库

### 使用Python访问MySQL数据库


  - `ipython-sql`：一个用于在IPython（Jupyter的核心）中直接运行SQL语句的扩展库。它允许用户通过`%sql`或`%%sql`魔术命令（Magic Commands）直接在代码单元中编写SQL，并连接数据库执行查询。

  - `pymysql`：实现了MySQL客户端/服务器协议，允许Python程序直接与MySQL数据库交互（执行SQL、管理连接等）。它无需依赖MySQL原生驱动，兼容性和安装便捷性较好。

- **用途**：  
  提供Python与MySQL数据库之间的连接能力，常用于从Python脚本或Jupyter中操作MySQL数据库。

```python
%load_ext sql
%config SqlMagic.autocommit=False
# 设置不同的表格样式
from prettytable import MARKDOWN
%config SqlMagic.style = 'MARKDOWN'   # 或其他可用样式，如 'MSWORD_FRIENDLY', 'PLAIN_COLUMNS' 等
%sql mysql+pymysql://root:12345678@localhost
```

```python
%%sql
SHOW DATABASES;
```

```python
%%sql
DROP TABLE IF EXISTS companies;
CREATE TABLE companies (
    id INT UNSIGNED AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    industry VARCHAR(100),
    email VARCHAR(100),
    phone VARCHAR(20),
    address TEXT,
    city VARCHAR(100),
    country VARCHAR(100) DEFAULT '中国',
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    is_active BOOLEAN DEFAULT TRUE
);
```
